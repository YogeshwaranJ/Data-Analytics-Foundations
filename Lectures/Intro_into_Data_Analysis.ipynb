# Part 1: Data Analytics Fundamentals

## 1.1 What is Data Analysis?

At its core, **Data Analysis** is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making.

**Why is it so important?**

We live in a world of data. Every click, every purchase, every GPS location... it's all data. Businesses, non-profits, and governments that can make sense of this data have a massive advantage.

Data Analysis helps us move from **data** to **information** to **insight** to **action**.



**Real-World Examples:**
* **Netflix:** Analyzes your viewing history to recommend what you should watch next.
* **Retail:** A grocery store analyzes purchase data to decide where to place items (e.g., putting beer next to diapers).
* **Healthcare:** Researchers analyze clinical trial data to determine if a new drug is effective and safe.

## 1.2 The Data Analysis Lifecycle

You don't just "do analysis." It's an iterative process, often called a lifecycle. A popular model is the 6-step process:



1.  **Ask (Problem Definition):** What business question are we trying to answer? This is the most important step. A good question leads to good insights. (e.g., "Why are our sales down in the Northeast region?")

2.  **Prepare (Data Collection):** Where do we get the data to answer this question? (e.g., Download sales reports from the database, get location data from a CSV, use a public API for weather data).

3.  **Process (Data Cleaning):** This often takes up to 80% of an analyst's time! Data in the real world is messy. We must handle:
    * Missing values (e.g., an 'age' field that is empty).
    * Duplicates (e.g., the same customer listed twice).
    * Incorrect data types (e.g., a 'date' stored as text).
    * Outliers (e.g., a product with a price of $9,999,999).

4.  **Analyze (EDA & Modeling):** This is the "fun" part. We explore the data, look for patterns, and test our hypotheses. This is often called **Exploratory Data Analysis (EDA)**. We might also build statistical or machine learning models.

5.  **Share (Visualization & Communication):** Data doesn't speak for itself. We must tell a story. We create charts, graphs, and dashboards (using tools like Matplotlib, Seaborn, or Tableau) and present our findings in a clear, compelling way.

6.  **Act (Decision Making):** The final step! The business uses our insights to make a decision. (e.g., "Our analysis shows the new competitor in the Northeast is the cause. We recommend a targeted marketing campaign to counter them.")

## 1.3 The 4 Types of Data Analytics

Analytics can be broken down into four main types, which build on each other in value and complexity.

[Image of the four types of data analytics (Descriptive, Diagnostic, Predictive, Prescriptive)]

1.  **Descriptive Analytics: "What happened?"**
    * This is the simplest form. It summarizes past data.
    * **Example:** A dashboard showing total sales for the last quarter. ($10 Million in sales).

2.  **Diagnostic Analytics: "Why did it happen?"**
    * This digs deeper to find the cause of what happened.
    * **Example:** Drilling into the sales data to find that sales were high *because* a new marketing campaign in the Northeast was very successful.

3.  **Predictive Analytics: "What will happen?"**
    * This uses historical data to forecast future outcomes.
    * **Example:** Using a model to forecast that sales for the *next* quarter will be $11 Million.

4.  **Prescriptive Analytics: "What should we do?"**
    * This is the most complex. It recommends actions to take to achieve a desired outcome.
    * **Example:** A model recommends that to achieve $12 Million in sales, the company should increase the marketing budget in the Western region by 15%.

## 1.4 Practical Lab: Exploring the Titanic Dataset

Now, let's go through a mini-analysis lifecycle (Ask, Prepare, Process, Analyze, Share) using Python.

**Our Toolset:**
* **Python:** A powerful, easy-to-read programming language.
* **Pandas:** The most important Python library for data manipulation and analysis. Think of it as Excel on steroids, but in code.
* **Matplotlib & Seaborn:** Python libraries for data visualization (making charts and graphs).

````python
# First, we need to import our libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# This line makes our plots appear inside the notebook
%matplotlib inline

# This sets a nice default style for our plots
sns.set_style('whitegrid')

print("Libraries imported successfully!")
````

### Step 1 & 2: Ask & Prepare (Load Data)

**Question:** What factors influenced a person's survival on the Titanic?

**Prepare:** We will load a famous dataset that comes built-in with the Seaborn library. A `DataFrame` is the main object in Pandas - it's basically a table.

````python
# Seaborn has a few datasets built-in. Let's load the 'titanic' dataset.
df = sns.load_dataset('titanic')

print("Titanic dataset loaded.")
````

### Step 3: Process (Explore & Clean Data)

Let's see what our data looks like. We'll check for missing values and get a summary.

````python
# Look at the first 5 rows of our data
df.head()
````

````python
# Get a high-level summary of all columns: data types, non-null counts
df.info()
````

**Cleaning Insights from `.info()`:**

* We have 891 entries.
* `age` has only 714 non-null values. That's 177 missing!
* `embarked` has 889 non-null values (2 missing).
* `deck` has only 203 non-null values. It's mostly empty!
* `embarked_town` also has 2 missing.

**Let's clean this up:**
1.  We'll fill the missing `age` values with the average (median) age. This is a common technique called **imputation**.
2.  The `deck` column is too empty to be useful. Let's **drop** (delete) it.
3.  Since only 2 `embarked` values are missing, let's just drop those rows.

````python
# 1. Impute 'age' with the median age
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)

# 2. Drop the 'deck' column (axis=1 means drop a column)
df_cleaned = df.drop('deck', axis=1)

# 3. Drop the few rows with missing 'embarked' data
df_cleaned = df_cleaned.dropna(subset=['embarked', 'embark_town'])

# Let's check our work!
df_cleaned.info()
````

Much better! Now we have 889 passengers and no missing values in the columns we care about.

### Step 4 & 5: Analyze & Share (Visualize)

Now for the fun part. Let's answer our question by making some plots.

#### Analysis Question 1: How many people survived vs. did not survive?

````python
# We can use a simple count plot.
# 0 = No, 1 = Yes
sns.countplot(x='survived', data=df_cleaned)
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.show()
````

**Insight:** Far more people did not survive than survived.

#### Analysis Question 2: Did passenger class (`pclass`) affect survival?

````python
# We can use 'hue' to split the 'survived' bar by another category
sns.countplot(x='survived', hue='pclass', data=df_cleaned)
plt.title('Survival Count by Passenger Class')
plt.legend(title='Class', labels=['1st', '2nd', '3rd'])
plt.show()
````

**Insight:** Wow! This is a powerful chart.
* For those who **did not survive** (0), the vast majority were in 3rd class.
* For those who **did survive** (1), a very large portion were in 1st class.

This clearly suggests that being in 1st class gave you a much better chance of survival.

#### Analysis Question 3: Did gender (`sex`) affect survival?

````python
sns.countplot(x='survived', hue='sex', data=df_cleaned)
plt.title('Survival Count by Gender')
plt.show()
````

**Insight:** Another very strong finding.
* The majority of those who **did not survive** (0) were male.
* The majority of those who **did survive** (1) were female.

This points to the "women and children first" protocol.

#### Analysis Question 4: What was the age distribution of survivors?

````python
# Let's use a histogram (distplot or histplot) to see the age distributions
# We can overlay the two plots for 'survived' vs 'not survived'

g = sns.FacetGrid(df_cleaned, col='survived')
g.map(sns.histplot, 'age', bins=20, kde=True)
plt.show()
````

**Insight:**
* The 'survived=0' (left) chart shows a large spike for people in their 20s-30s (the main workforce).
* The 'survived=1' (right) chart shows a significant spike for very young children.

This confirms the "women and **children** first" idea.

### Part 1 Conclusion

Look at what we just did! We took a raw, messy dataset, asked a question, cleaned the data, and then created visualizations that gave us clear, actionable insights.

**Final Answer:** A person's chance of survival was *highly* dependent on their **class**, **gender**, and **age**.

# Part 2: Basics of Structured Query Language (SQL)

## 2.1 What is SQL?

SQL (often pronounced "sequel") stands for **Structured Query Language**.

**Wait, why do we need SQL if we have Pandas?**

* **Pandas** is great for data that fits on your computer (like a CSV or Excel file).
* **SQL** is the language used to communicate with **Relational Databases**. Databases are designed to store, manage, and retrieve *massive* amounts of data efficiently and safely on a server.

Think of it this way: Pandas is like your personal notebook. A database is like a giant, secure, organized library for an entire company. SQL is the language you use to ask the librarian to get you a specific book.

### Relational Databases

These databases store data in **tables** (like our Titanic `DataFrame`). These tables can be related to each other.



* **Tables:** Store data about an entity (e.g., `Customers`, `Products`).
* **Rows:** A single record (e.g., one customer, one product).
* **Columns:** A specific attribute (e.g., `FirstName`, `Price`).
* **Primary Key (PK):** A *unique* ID for each row in a table (e.g., `CustomerID`).
* **Foreign Key (FK):** A column in one table that refers to the Primary Key in another table. This is how we link them! (e.g., `CustomerID` in the `Orders` table).

## 2.2 The Sub-Languages of SQL (Theory)

SQL is often broken into 3-4 main parts. We'll focus on the theory here.

### 1. DDL (Data Definition Language)
This part of SQL defines the *structure* of the database. It's like building the empty library shelves.

* `CREATE`: Creates a new table or database.
    * `CREATE TABLE Users (UserID int, LastName varchar(255));`

* `ALTER`: Modifies an existing table (e.g., add or remove a column).
    * `ALTER TABLE Users ADD Email varchar(255);`

* `DROP`: Deletes an entire table. (Very permanent!)
    * `DROP TABLE Users;`

### 2. DML (Data Manipulation Language)
This is the most common part. It's used to interact with the *data* inside the tables. It's like adding, removing, or reading the books on the shelves.

* `SELECT`: **The most important command.** Used to retrieve data.
    * `SELECT * FROM Users;`

* `INSERT`: Adds new rows (records) to a table.
    * `INSERT INTO Users (UserID, LastName) VALUES (1, 'Smith');`

* `UPDATE`: Modifies existing rows in a table.
    * `UPDATE Users SET LastName = 'Jones' WHERE UserID = 1;`

* `DELETE`: Deletes rows from a table.
    * `DELETE FROM Users WHERE UserID = 1;`

### 3. DCL (Data Control Language)
This part manages permissions and access. It's the "security guard" of the database.

* `GRANT`: Gives a specific user permission to do something (e.g., `SELECT` data).
    * `GRANT SELECT ON Users TO 'some_user';`

* `REVOKE`: Takes away a permission.
    * `REVOKE SELECT ON Users FROM 'some_user';`

## 2.3 Deep Dive: The `SELECT` Statement

As an analyst, you will spend 95% of your time using `SELECT`. Let's learn its main clauses.

Imagine we have a table called `Products`:

| ProductID | ProductName | Category | Price |
|---|---|---|---|
| 1 | Laptop | Electronics | 1200 |
| 2 | Coffee | Food | 15 |
| 3 | Chair | Furniture | 85 |
| 4 | Mouse | Electronics | 40 |
| 5 | Bread | Food | 4 |

---
### `SELECT` and `FROM`
Tells the database *what columns* you want from *which table*.

```sql
/* Get all columns (using *) */
SELECT * FROM Products;

/* Get only specific columns */
SELECT ProductName, Price FROM Products;
```

---
### `WHERE` (Filtering)
Filters the rows to only get the ones that match a condition.

```sql
/* Find all products in the 'Electronics' category */
SELECT * FROM Products
WHERE Category = 'Electronics';

/* Find all products that cost more than $50 */
SELECT * FROM Products
WHERE Price > 50;
```
**Result:** (Laptop, Chair, Mouse)

---
### Aggregate Functions & `GROUP BY`
**Aggregate functions** perform a calculation on a set of rows and return a single value.
Common ones: `COUNT()`, `SUM()`, `AVG()`, `MIN()`, `MAX()`.

**`GROUP BY`** is used with aggregate functions to group the result-set by one or more columns.

```sql
/* How many products do we have? */
SELECT COUNT(*) FROM Products;
/* Result: 5 */

/* What is the average price of all products? */
SELECT AVG(Price) FROM Products;
/* Result: 268.8 */

/* This is the powerful one: What is the average price PER category? */
SELECT Category, AVG(Price)
FROM Products
GROUP BY Category;
```
**Result:**
| Category | AVG(Price) |
|---|---|
| Electronics | 620 |
| Food | 9.5 |
| Furniture | 85 |

---
### `ORDER BY` (Sorting)
Sorts the final result. Default is ascending (`ASC`). You can use `DESC` for descending.

```sql
/* List all products from most expensive to least expensive */
SELECT * FROM Products
ORDER BY Price DESC;
```
**Result:** (Laptop, Chair, Mouse, Coffee, Bread)

## 2.4 `JOIN`s (The Core Concept)

The real power of SQL comes from combining tables. This is done with a `JOIN`.

Let's add a second table, `Orders`:
| OrderID | ProductID | Quantity |
|---|---|---|
| 101 | 2 | 3 |
| 102 | 1 | 1 |
| 103 | 4 | 2 |

**Problem:** If I look at the `Orders` table, I only see `ProductID` 2. I have no idea what that product is. I need to `JOIN` it with the `Products` table on the common key (`ProductID`).



An **`INNER JOIN`** gets rows *only* where there is a match in *both* tables.

```sql
/* Show me the product name for each order */
SELECT O.OrderID, P.ProductName, O.Quantity
FROM Orders AS O
INNER JOIN Products AS P
ON O.ProductID = P.ProductID;
```
(Note: `AS O` and `AS P` are just short aliases to make the query easier to write.)

**Final Result of the `JOIN`:**
| OrderID | ProductName | Quantity |
|---|---|---|
| 101 | Coffee | 3 |
| 102 | Laptop | 1 |
| 103 | Mouse | 2 |

We have now combined two tables to get a much more useful result!
